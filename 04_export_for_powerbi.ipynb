```python
import pandas as pd
import sqlite3
import os

DB_PATH     = 'C:/Users/Saurabh yadav/Documents/SaaS Growth Intelligence Platform/database/saas_data.db'
EXPORT_PATH = 'C:/Users/Saurabh yadav/Documents/SaaS Growth Intelligence Platform/powerbi/'

os.makedirs(EXPORT_PATH, exist_ok=True)

conn = sqlite3.connect(DB_PATH)

tables = ['accounts', 'subscriptions', 'churn_events',
          'feature_usage', 'support_tickets', 'churn_scores', 'mrr_forecast']

for table in tables:
    df = pd.read_sql('SELECT * FROM ' + table, conn)
    df.to_csv(EXPORT_PATH + table + '.csv', index=False)
    print('Exported:', table, '—', len(df), 'rows')

conn.close()
print('All files exported to powerbi folder')
```
